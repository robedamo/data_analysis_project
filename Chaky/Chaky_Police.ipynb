```python
#!/usr/bin/env python

# make sure to install these packages before running:
# pip install pandas
# pip install sodapy

import pandas as pd
from sodapy import Socrata

# Unauthenticated client only works with public data sets. Note 'None'
# in place of application token, and no username or password:
client = Socrata("analisi.transparenciacatalunya.cat", None)

# Example authenticated client (needed for non-public datasets):
# client = Socrata(analisi.transparenciacatalunya.cat,
#                  MyAppToken,
#                  username="user@example.com",
#                  password="AFakePassword")

# First 2000 results, returned as JSON from API / converted to Python list of
# dictionaries by sodapy.
results = client.get("gci6-2ubm", limit=2000)

# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)
```

```python
dic_1est = {}
k = 0
while k < len(var.index):
    edat = var.index[k][0]
    v_o_pa = var.index[k][1]
    if edat in dic_1est:
        dic_1est[edat][v_o_pa] = var[k]
    else:
        dic_1est[edat] = {v_o_pa : var[k]}
    k += 1
    
dic_1est
```

```python
ordenar = []
for i in dic_1est:
    ordenar.append(int(i))
    
ordenar.sort()
#ordenar

dic_1est_def = {}

for i in ordenar:
    dic_1est_def[i] = dic_1est[str(i)]
    
dic_1est_def
print(len(dic_1est_def))
```

```python
import matplotlib.pyplot as plt
import numpy as np


labels = []
for i in dic_1est:
    labels.append(i)
pa = []
for i in dic_1est:
    try:
        pa.append(dic_1est[i]['Persona autora'])
    except:
        pa.append(0)

v = []
for i in dic_1est:
    try:
        v.append(dic_1est[i]['Víctima'])
    except:
        v.append(0)
        
#print(pa,v)
#print(len(pa),len(v),len(labels))
#width = 0.35       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

ax.bar(labels, pa, label='Persona autora')
ax.bar(labels, v, bottom=pa, label='Víctima')

ax.set_ylabel('Nombre de casos')
ax.set_title('Edats')
ax.legend()
plt.xticks(np.arange(0, 90, step=5))

plt.show()
```

```python
#hago lo mismo ahora per con porcentajes
y_nova = []
i = 0
while i < len(labels):
    y_nova.append(v[i]/(pa[i]+v[i]))
    i += 1
    
len(y_nova)
plt.scatter(labels,y_nova)
plt.xticks(np.arange(0, 90, step=5))
```
