```python
#!/usr/bin/env python

# make sure to install these packages before running:
# pip install pandas
# pip install sodapy

import pandas as pd
from sodapy import Socrata

# Unauthenticated client only works with public data sets. Note 'None'
# in place of application token, and no username or password:
client = Socrata("analisi.transparenciacatalunya.cat", None)

# Example authenticated client (needed for non-public datasets):
# client = Socrata(analisi.transparenciacatalunya.cat,
#                  MyAppToken,
#                  username="user@example.com",
#                  password="AFakePassword")

# First 2000 results, returned as JSON from API / converted to Python list of
# dictionaries by sodapy.
results = client.get("gci6-2ubm", limit=2000)

# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)
```

```python
llista_categories
dic_df = {}
for i in llista_categories:
    boludo = results_df[i].value_counts()
    values_categoria = []
    for k in boludo:
        values_categoria.append(k)    
    #values_any

    #boludo.index
    respostes_categoria = []
    for k in boludo.index:
        respostes_categoria.append(k)

    dic_categoria = {}
    k = 0
    while k < len(values_categoria):
        dic_categoria[respostes_categoria[k]] = values_categoria[k]
        k += 1
        
    try:
        respostes_categoria2 = []
        for k in respostes_categoria:
            k = int(k)
            respostes_categoria2.append(k)
        respostes_categoria2.sort()
            
    except:
        respostes_categoria2 = respostes_categoria

    #print(respostes_categoria2)
    dic_categoria2 = {}
    for k in respostes_categoria2:
        #print(k)
        for j in dic_categoria:
            try:
                a = int(j)
            except:
                a = j
            #print(j)
            if a == k:
                dic_categoria2[k] = dic_categoria[j]
            #print(dic_categoria2)
        
    dic_df[i] = dic_categoria2
    
dic_df
```

```python
import matplotlib.pyplot as plt

for i in llista_categories:
    dictionary = dic_df[i]
    plt.bar(list(dictionary.keys()), dictionary.values(), )#color='g')
    plt.title(i)
    plt.savefig(i+'_total.png')
    plt.show()
```
